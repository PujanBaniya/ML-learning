```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('Sex',OneHotEncoder(drop = 'first'),['Sex'])],remainder = 'passthrough')
```

```python
train
```

```python
train_array = ct.fit_transform(train)
```

```python
train_array[:,:1]
```

```python
columns= ct.get_feature_names_out()
```

```python
 ct.get_feature_names_out()
```

```python
df = pd.DataFrame(train_array, columns = columns)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df.drop(['remainder__id'],axis= 1,inplace = True)
```

```python
df
```

```python
X= df.drop(['remainder__Calories'],axis =1)
y= df['remainder__Calories']
```

```python
y
```

```python
X_train,X_val,y_train,y_val = train_test_split(X,y, random_state= 42, test_size = 1/3)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_val)
print("Mean_absolute_erroe:", mean_absolute_error(y_val, y_pred))
print("MeanSquareError:", mean_squared_error(y_val, y_pred))
print("RootMeanSE:", np.sqrt(mean_squared_error(y_val, y_pred)))
print("R2 Score:", r2_score(y_val, y_pred))
```

```python
test_predict = model.predict(X_val)
```

```python
submission = pd.DataFrame()
submission['id']= test['id']
submission['Calories'] = test_predict
submission.to_csv('my_submission.csv', index=False)
```
